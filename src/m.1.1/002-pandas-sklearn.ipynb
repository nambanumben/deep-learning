## Pandas and sklearn

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python

# Read in csv file
# File: https://github.com/mGalarnyk/Python_Tutorials/blob/master/Python_Basics
raw_data = pd.read_csv("https://raw.githubusercontent.com/mGalarnyk/Python_Tutorials/refs/heads/master/Python_Basics/Linear_Regression/linear.csv")
```

```python

# Removes rows with NaN in them
filtered_data = raw_data[~np.isnan (raw_data["y"])]

x_y = np.array (filtered_data)
x, y=x_y[:,0], x_y[:,1]

# Reshaping
x, y=x.reshape (-1,1), y.reshape(-1, 1)
```

```python

# Linear Regression Object
lin_regression = LinearRegression()

# Fitting linear model to the data
# h(x)= theta0 x0 + thetal x1
lin_regression.fit(x,y)

# Get slope of fitted line
theta1 = lin_regression.coef_

# Get y-Intercept of the Line
theta0 = lin_regression.intercept_

# Get Predictions for original x values
# you can also get predictions for new data
predictions = lin_regression.predict(x)
```

```python


# following slope intercept form
#print('Hello, my name is + name + and I am + str(age))
#print('My name is {name} and I am {age}'.format(name-name, age-age))
print ('formula: h(x) = '+str(theta0)+'x0 + '+str(theta1)+' x1')
```

```python

# Plot the Original Model (Black) and Predictions (Blue)
plt.scatter(x, y, color='red')
plt.plot(x, predictions, color='blue', linewidth=3) 
plt.show()
```
